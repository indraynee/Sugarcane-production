# data clenaing

```python
df["Production (Tons)"] = df["Production (Tons)"].str.replace(".","")
df["Production per Person (Kg)"] = df["Production per Person (Kg)"].str.replace(".","").str.replace(",",".")
df["Acreage (Hectare)"] = df["Acreage (Hectare)"].str.replace(".","")
df["Yield (Kg / Hectare)"]= df["Yield (Kg / Hectare)"].str.replace(".","").str.replace(",",".")
```

```python
df
```

```python
df.isna().sum()
```

```python
df[df["Acreage (Hectare)"].isnull()]
```

```python
df = df.dropna().reset_index().drop("index", axis = 1)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.drop(["Unnamed: 0"],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df["Production (Tons)"] = pd.to_numeric(df["Production (Tons)"], errors='coerce')
df["Production per Person (Kg)"] = pd.to_numeric(df["Production per Person (Kg)"], errors='coerce')
df["Acreage (Hectare)"] = pd.to_numeric(df["Acreage (Hectare)"], errors='coerce')
df["Yield (Kg / Hectare)"] = pd.to_numeric(df["Yield (Kg / Hectare)"], errors='coerce')
```

## Do number of countries in a continent affects production in sugarcane ?


```python
df_continent["number_of_countries"]=df.groupby("Continent").count()["Country"]
df_continent
```

```python
continents_name=df_continent.index.to_list()
sns.lineplot(data=df_continent,x="number_of_countries",y="Production (Tons)")
plt.xticks(df_continent["number_of_countries"],continents_name,rotation=90)
plt.show()
```

##### the  number of countries in a continent affects does not affect production in sugarcane
